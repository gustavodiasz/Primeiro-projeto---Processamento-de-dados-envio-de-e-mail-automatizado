# ENVIANDO E-MAILS AUTOMATICAMENTE COM O PYTHON

## ETAPAS DO PROJETO

1) Importar base de dados;

2) Visualisar base e modificar se nescessário;

3) Faturamento por loja;

4) Quantidade de produtos vendidos por loja;

5) Ticket médio por produto em cada loja;

6) Enviar um e-mail com relatório.

## BIBLIOTECA(AS) NESCESSÁRIA(AS)

   PANDAS (manilular dados do Excel)
   
   win32com.client (mandar e-mails)


```python
# IMPORTANDO E VISUALIZANDO

import pandas as pd

vendas_tabela = pd.read_excel('Vendas-MiniCurso-Hastag-projeto.xlsx')

vendas_tabela
```

## REAIZANDO O FATURAMENTO POR LOJA

```python
faturamento = vendas_tabela[['ID Loja','Valor Final']].groupby('ID Loja').sum()

faturamento
```

## REAIZANDO A QUANTIDADE DE PRODUTOS VENDIDOS POR LOJA

```python
quantidade = vendas_tabela[['ID Loja','Quantidade']].groupby('ID Loja').sum()

quantidade
```

## REALIZANDO O TICKET MÉDIO POR LOJA

```python
# ticket é faturamento/quantidade 

# para a divisão das tabelas retornarem no formato de tabela deve ser aplicado o comando (código da dividão vem aqui dentro).to_frame()

ticket_médio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_médio = ticket_médio.rename(columns={0:'ticket médio'})

ticket_médio
```

## ENVIAR E-MAIL AUTOMATICAMENTE

```python
# esse trecho de código não precisa decorar!

# O corpo do e-mail esta em HTMLBody. Nesse tipo de linguegem os paragrafos são identificados quando estão entre: <p> paragrafo </p>

# Para o envio dos e-mails deve-se ter o Outlook instalado e logado.

import win32com.client as win32

outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)
mail.To = 'gustavod.junior@gmail.com'
mail.Subject = 'RELATÓRIO DE VENDAS POR LOJA'
mail.HTMLBody = f'''

<p>Prezados colabradores,</p> 

<p>Segue tabelas dos relatório de vendas com: Faturamento por loja, Quantidade de produtos vendidos por loja, Ticket médio por produto em cada loja.</p>


<p>Faturamento por loja:</p>
{faturamento.to_html(formatters={'Valor Final':'R${:,.2f}'.format})}

<p>Quantidade de produtos vendidos por loja:</p>
{quantidade.to_html(formatters={'Quantidade':'R${:,.2f}'.format})}

<p>Ticket médio por produto em cada loja:</p>
{ticket_médio.to_html(formatters={'ticket médio':'R${:,.2f}'.format})}

<p>Favor acusar recebiento.</p>

<p>Cordialmente,</p>

<p>Gustavo Dias</p>

'''

mail.Send()
```
